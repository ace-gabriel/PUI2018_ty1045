```python
import numpy as np
import scipy as sp
import statsmodels.formula.api as smf
import pylab as pl
import pandas as pd
%pylab inline
```

```python
df = pd.read_csv('Asc.csv', header=None, names=['x', 'y1','tmp2', 'y2', 'tmp3', 'tmp4', 'tmp5', 'tmp6'], sep="\t")
```

```python
df
```

```python
df.sort_values(by='x', inplace=True)
```

```python
df
```

```python
pl.figure(figsize=(5,5))
pl.plot(df['x'], df['y1'], 'go')
pl.plot(df['x'], df['y2'], 'r*')
pl.xlabel('Independent')
pl.ylabel('Depedent')
pl.title('Anscomby 1 & 2');
```

# First Anscomby distribution: y1

## OLS linear fit  y1~x

```python
OLS1 = smf.ols(formula='y1  ~ x', data=df).fit()
OLS1.summary()
```

```python
pl.figure(figsize=(5,5))
pl.plot(df['x'],df['y1'],'go')
pl.plot(df['x'], OLS1.predict(df),'-', color='orange')
pl.xlabel('Independent')
pl.ylabel('Depedent')
pl.title('Anscomby 1 - line fit');
```

## OLS quadratic fit  y1~x^2 + x

```python
# create a new feature for x^2
df['x2'] = df['x']**2
```

```python
OLS2 = smf.ols(formula='y1 ~ x2 + x',data=df).fit()
OLS2.summary()
```

```python
pl.figure(figsize=(5,5))
pl.plot(df['x'],df['y1'],'go')
pl.plot(df['x'], OLS2.predict(df),'-', color='orange')
pl.xlabel('Independent')
pl.ylabel('Depedent')
pl.title('Anscomby 1 - quadratic fit');
```

# Compare the models with the LR ratio test

```python
OLS2.compare_lr_test(OLS1)
```

```python
print("Likelihood ratio: %f" % (-2 * (OLS1.llf - OLS2.llf)))
```

### H0: More complex model does not fit better compared to the simpler one (restricted model)

### Result Explanation:
#### 1) First value is the likelyhood ratio, it equals -2 * (OLS1.llf - OLS2.llf). It is the chisquare distributed with df_diff degrees of freedom
#### 2) The second one is the p-value of the test statistic.
#### 3) The third one is the degree of freedom in which the restricted model follows.
#### Moreover, the significant value under degree of freedom of 1 and alpha of 0.05 is 3.84 and 3.84 > 0.7079. Moreover, t statistic is 0.4, and 0.4 > 0.05. We can't reject the null hypothesis that more complex model better fits the data.

Equation of LR:
    
    LR = -2 log (Likelihood_restricted / likelihood_complex)
    
or equivalently

    LR = -2 (log(Likelihood)_restricted - log(Likelihood)_complex)

where log is the natural log (base e)
